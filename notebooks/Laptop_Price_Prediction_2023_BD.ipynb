#### adding the SD,HD,FHD,2K,4K,Retina columns

```python
resolution_dict = {
    '2560x1600': '2K',
    '2880x1800': '2K',
    '1920x1080': 'FHD',
    '3456x2234': '2K',
    '3024x1964': '2K',
    '2560x1440': '2K',
    '3840x2160': '4K',
    '3840x2400': '4K',
    '2240x1400': '2K',
    '1920x1200': 'FHD',
    '2560x1920': '2K',
    '1366x768': 'HD',
    '1280x720': 'HD',
    '2560x1664': '2K',
    '1360x768': 'HD',
    '366x768': 'SD',
    '1280x800': 'HD',
    '1920x1280': 'FHD',
}
```

```python
df['SD'] = 0
df['HD'] = 0
df['FHD'] = 0
df['2K'] = 0
df['4K'] = 0

def display(val):
    category = resolution_dict[val]
    df.loc[df['sc_res1'] == val, category] = 1
```

#### Adding refresh rate column

```python
df[df['ScreenResolution'].apply(lambda x: '165hz' in x.lower().replace(' ',''))]['ScreenResolution'].tolist()
```

```python
df['Refresh Rate'] = 60

def refresh_rate(x):
    info = x.lower().replace(' ', '')
    if '60hz' in info:
        df.loc[df['ScreenResolution'] == x, 'Refresh Rate'] = 60
    elif '120hz' in info:
        df.loc[df['ScreenResolution'] == x, 'Refresh Rate'] = 120
    elif '144hz' in info:
        df.loc[df['ScreenResolution'] == x, 'Refresh Rate'] = 144
    elif '165hz' in info:
        df.loc[df['ScreenResolution'] == x, 'Refresh Rate'] = 165
    elif '240hz' in info:
        df.loc[df['ScreenResolution'] == x, 'Refresh Rate'] = 240
    elif '300hz' in info:
        df.loc[df['ScreenResolution'] == x, 'Refresh Rate'] = 300
    else:
        df.loc[df['ScreenResolution'] == x, 'Refresh Rate'] = 60
        
        
```

```python
df['ScreenResolution'].apply(refresh_rate)
```
